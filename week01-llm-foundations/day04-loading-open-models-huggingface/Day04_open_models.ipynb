```python
from transformers import pipeline
import matplotlib.pyplot as plt

print("=== Day 4 — Loading Open Models with Hugging Face ===\n")

prompt = "My firewall keeps blocking"
```

## 1. Base Model (GPT-2)

```python
generator = pipeline("text-generation", model="gpt2")

temperatures = [0.3, 0.7, 1.2]
generated_texts = []

for temp in temperatures:
    result = generator(prompt, max_new_tokens=40, temperature=temp, do_sample=True, pad_token_id=50256)
    text = result[0]["generated_text"]
    generated_texts.append(text)
    print(f"\nTemperature = {temp}")
    print(text)
    print("-" * 80)
```

## 2. Instruction-Tuned Model

```python
chat = pipeline("text-generation", model="Qwen/Qwen2.5-0.5B-Instruct")

messages = [{"role": "user", "content": "Why does a firewall sometimes block legitimate traffic?"}]

response = chat(messages, max_new_tokens=120, temperature=0.7)

print("Question:", messages[0]["content"])
print("\nAnswer:")
print(response[0]["generated_text"][-1]["content"])
```

## 3. Temperature Effect Visualization

```python
plt.figure(figsize=(12, 9))
for i, temp in enumerate(temperatures):
    plt.subplot(len(temperatures), 1, i+1)
    plt.text(0.5, 0.5, generated_texts[i], ha='center', va='center', fontsize=11, wrap=True)
    plt.title(f"Temperature = {temp}", fontsize=14)
    plt.axis('off')
plt.suptitle("Effect of Temperature on Text Generation (GPT-2)", fontsize=16)
plt.tight_layout()
plt.show()
```
