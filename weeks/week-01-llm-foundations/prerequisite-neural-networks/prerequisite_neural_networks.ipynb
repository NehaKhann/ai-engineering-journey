# Prerequisite — Neural Networks Fundamentals

Understand the basic building blocks of neural networks before diving into Transformers and LLMs.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

## 1. What is a Neural Network?

```python
print("A simple neural network is like a function that learns from data.")
print("It takes input → processes through layers → gives output.")
print("During training, it adjusts its weights to make better predictions.\n")
```

## 2. Creating a Simple Neural Network

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(in_features=2, out_features=1)  # 2 inputs → 1 output
    
    def forward(self, x):
        return self.linear(x)

model = SimpleNN()
print("Simple Neural Network created!")
print(model)
```

## 3. Forward Pass (Making Predictions)

```python
# Example input: 2 samples, each with 2 features
x = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
print("Input:", x)

output = model(x)
print("\nPrediction (Output):", output)
```

## 4. Training Loop (How the Model Learns)

```python
# Target: We want the model to output 5.0 for input [2.0, 3.0]
target = torch.tensor([[5.0]])

loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

losses = []

print("Starting Training...\n")

for epoch in range(15):
    # Forward Pass
    prediction = model(x)
    loss = loss_fn(prediction[0], target)
    losses.append(loss.item())
    
    # Backward Pass + Update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if epoch % 3 == 0:
        print(f"Epoch {epoch:2d} | Loss: {loss.item():.4f} | Prediction: {prediction[0].item():.4f}")

print("\nTraining Finished!")
```

## 5. Loss Curve Visualization

```python
plt.figure(figsize=(8, 5))
plt.plot(losses, marker='o', color='blue')
plt.title("Loss Curve - Model is Learning")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

## Key Takeaways

- Neural Networks learn by adjusting **weights** and **bias**
- The learning process is: **Forward Pass → Loss → Backward Pass → Update**
- PyTorch makes this easy with `nn.Module`, `loss.backward()`, and optimizers
- This same idea scales up to massive models like GPT
